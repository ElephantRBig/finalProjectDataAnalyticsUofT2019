```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'allMarketsConsolidatedHereVersion2.csv')
canDesc = pd.read_csv(r'CANADA2019Jan1-2019Nov4CustomTransaction.csv')
usDesc = pd.read_csv(r'USA2019Jan1-2019Nov4CustomTransaction (2).csv')
netDesc = pd.read_csv(r'itemDescription.csv')
```

```python
data['item description'] = ''
```

```python
# adding the extra M for canDesc and US
# Inserting an extra 'M' into all skus which have the first 3 characters '3M-...'

canDesc['sku'] =canDesc['sku'].astype(str)
usDesc['sku'] =usDesc['sku'].astype(str)

for i in range(0,len(canDesc['sku'])):
    if canDesc['sku'][i][0:3] == '3M-':
        canDesc['sku'][i] = canDesc['sku'][i][0:2]+'M'+canDesc['sku'][i][2:len(canDesc['sku'][i])]

for i in range(0,len(usDesc['sku'])):
    if usDesc['sku'][i][0:3] == '3M-':
        usDesc['sku'][i] = usDesc['sku'][i][0:2]+'M'+usDesc['sku'][i][2:len(usDesc['sku'][i])]
```

```python
canDesc = canDesc.iloc[:,[5,6]]
usDesc = usDesc.iloc[:,[4,5]]
netDesc['SKU'] = netDesc['SKU'].str.upper()
```

```python
canDesc['sku'] = canDesc['sku'].str.upper()
usDesc['sku'] = usDesc['sku'].str.upper()
```

```python
# azn Can description
canDescUnique = canDesc.groupby('sku').first()
canDescUnique = canDescUnique.reset_index()
canDescUnique.head()
```

```python
# azn US description
usDescUnique = usDesc.groupby('sku').first()
usDescUnique = usDescUnique.reset_index()
usDescUnique.head()
```

```python
netDesc.head()
```

```python
itemDesclist = []
itemDesclist = pd.DataFrame(itemDesclist)
```

```python
itemDesclist['item'] = ''
itemDesclist['item description'] = ''
```

```python
#canada loopa
for i in range(0,len(canDescUnique['sku'])):
    data.loc[data['item']== canDescUnique['sku'][i], 'item description'] = canDescUnique['description'][i]
    
for i in range(0,len(usDescUnique['sku'])):
    data.loc[data['item']== usDescUnique['sku'][i], 'item description'] = usDescUnique['description'][i]
    
for i in range(0,len(netDesc['SKU'])):
    data.loc[data['item']== netDesc['SKU'][i], 'item description'] = netDesc['Description'][i]
```

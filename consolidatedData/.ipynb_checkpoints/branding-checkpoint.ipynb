```python
listOfUniqueItemsMCR = [] 
for i in range(0,len(data['item'])):
    if data['item'][i][0:3] == "MCR": 
        listOfUniqueItemsMCR.append(data['item'][i])
        
for i in listOfUniqueItemsMCR:
    data.loc[data['item']==i,"product brand"] = "Memphis Med-Tech"
```

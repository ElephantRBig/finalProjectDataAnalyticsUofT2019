```python
import pandas as pd
import numpy as np
```

```python
net = pd.read_csv(r'net32ReportForConsolidation.csv')
zips = pd.read_csv(r'uszipS.csv',encoding = "ISO-8859-1")
```

```python
net.head()
```

```python
zips.head()
```

```python
net.loc[np.isnan(net["Ship Postal Code"]), 'Ship Postal Code'] = 0
net.loc[np.isnan(net["order city"]), 'order city'] = 0
net.loc[np.isnan(net["order state"]), 'order state'] = 0
```

```python
net['Ship Postal Code'] = net['Ship Postal Code'].astype('int64')
```

```python
net.loc[net['Ship Postal Code']=='77074']
```

```python
net.head()
```

```python
# for i in range(0,len(net['Ship Postal Code'].unique())):
for j in range(0,len(zips['zip'])):
    net.loc[net['Ship Postal Code'] == zips['zip'][j], 'order state']= zips['state_name'][j]
    net.loc[net['Ship Postal Code'] == zips['zip'][j], 'order city']= zips['city'][j]
```

```python
net = net.drop(columns = {'Ship Postal Code'})
net.head()
```

```python
net.to_csv('net32ConsolidatedReady.csv')
```

# fixing the states to be two letter 

```python
net = pd.read_csv('net32ConsolidatedReady.csv')
```

```python
state = pd.read_csv('provincestate.csv')
```

```python
net.head(1)
```

```python
for i in range(0,len(net['order state'].unique())):
    for j in range(0,len(state['State'])):
        if net['order state'].unique()[i] == state['State'][j]:
            net.loc[net['order state']==net['order state'].unique()[i], 'order state'] = state['Code'][j]
```

```python
net.to_csv('net32ConsolidatedReady1.csv')
```

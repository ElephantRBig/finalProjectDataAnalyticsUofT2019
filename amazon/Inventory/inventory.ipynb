```python
import pandas as pd 
import numpy as np
inventory = pd.read_csv("inventoryClean.csv")
inventory
```

```python
inventory = inventory.drop(columns = {"Unnamed: 0"})
```

```python
inventory = inventory.rename(columns = {'snapshot-date':'date'})
inventory = inventory.sort_values(by=['date'], ascending = True).reset_index().drop(columns = {'index'})
```

```python
# In this cell i am looking for the records that do not have a location/state
# for the fullfillment centre
missingIDs = inventory[inventory.isna().any(axis=1)]['fulfillment-center-id'].unique()
```

```python
# missing fullfillment centres and their locate
listOfMissingIds = []
for i in range(0,len(missingIDs)):
    listOfMissingIds.append(missingIDs[i])
```

```python
import pandas as pd
```

```python
facilityLocations = pd.read_csv("facilityLocations.csv")
inventory = pd.read_csv("amazonInventory.csv", encoding = "ISO-8859-1")
```

```python
inventory
```

```python
inventory['state'] = ''
inventory['location']=''
for i in range(0,len(inventory['fulfillment-center-id'].unique())):
    for j in range(0,len(facilityLocations['Code'].unique())):
        if inventory['fulfillment-center-id'].unique()[i] == facilityLocations['Code'].unique()[j]:
            inventory.loc[inventory['fulfillment-center-id']== inventory['fulfillment-center-id'].unique()[i],'state'] = facilityLocations['State'][j]
            inventory.loc[inventory['fulfillment-center-id']== inventory['fulfillment-center-id'].unique()[i],'location'] = facilityLocations['Location'][j]
            #inventory['location'][i] = facilityLocations['Location'][j]
            
# for i in range(0,len(usReport['sku'].unique())):
#     for j in range(0,len(database['Sub-ID'].unique())):
#             if test['sku'].unique()[i] == database['Sub-ID'][j]:
#                 test.loc[test['sku']== test['sku'].unique()[i],'group-sku'] = database['ID Group'][j]
```

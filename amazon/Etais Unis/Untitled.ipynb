```python
import pandas as pd
```

```python
orders = pd.read_csv(r'USorders.csv')
state = pd.read_csv(r'provincestate.csv')
```

```python
state['State'] = state['State'].str.lower()
orders['order state'] = orders['order state'].str.lower()
```

```python
# for i in range(0,len(orders['order state'].unique()):
list = []
for i in range(0,len(orders['order state'].unique())):
    if len(str(orders['order state'].unique()[i])) > 3:
        list.append(orders['order state'].unique()[i])
```

```python
orders['other'] = ''
```

```python
for i in range(0,len(list)):
    for j in range(0,len(state['State'])):
        if list[i] == state['State'][j]:
            orders.loc[orders['order state']==list[i], 'order state'] = state['Code'][j]
```

```python
orders['order state'] = orders['order state'].str.upper()
```

```python
orders = orders.drop(columns = {'other'})
```

```python
orders = orders.loc[orders['country']=='USA']
```

```python
ordersByState = orders.groupby('order state').agg({'profit':'sum'})
```

```python
ordersByState = ordersByState.reset_index()
```

```python
descendingProfitableStates = ordersByState.sort_values(by = 'profit', ascending = False).reset_index()
descendingProfitableStates = descendingProfitableStates.drop(columns = {'index'})
```

```python
descendingProfitableStates['State']=''
```

```python
for i in range(0,len(descendingProfitableStates['order state'])):
    for j in range(0,len(state['Code'])):
        if descendingProfitableStates['order state'][i] == state['Code'][j]:
            descendingProfitableStates['State'][i] = state['State'][j]
```

```python
for i in range(0,len(descendingProfitableStates['State'])):
    descendingProfitableStates['State'][i] = descendingProfitableStates['State'][i].capitalize()
```

# Adding the Return on sales Boolean

```python
import pandas as pd 
```

```python
y = pd.read_csv(r'aznUSAggregateReport.csv')
y
```

```python
y['1: R.O.S>=15%, 0: else'] = ''
for i in range(0,len(y['sku'])):
    if y['Return on Sales'][i] >= 15:
        y['1: R.O.S>=15%, 0: else'][i] = 1
    else:
        y['1: R.O.S>=15%, 0: else'][i] = 0
```

```python
y.to_csv(r'yROS.csv')
```

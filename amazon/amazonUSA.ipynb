```python
# I'm having an error with subscripting an iteration because of its type.
# Im going to change the type of the sku column from an object to a string
# so i can manipulate it
aznUS['sku'] =aznUS['sku'].astype(str)
```

```python
# Inserting an extra 'M' into all skus which have the first 3 characters '3M-...'
for i in range(0,len(aznUS['sku'])):
    if aznUS['sku'][i][0:3] == '3M-':
        aznUS['sku'][i] = aznUS['sku'][i][0:2]+'M'+aznUS['sku'][i][2:len(aznUS['sku'][i])]
```

```python
aznUS
```

```python
# In this cell, i wil be adding the main IDs first and then landing costs to each item by importing the 'databaseVersion1' 
databaseLandingCosts = pd.read_csv('azmDatabaseVersion1.csv')
```

```python
databaseLandingCosts.head(1)
```

```python
aznUS 
```

```python
# Here I am inserting a new column in between 'type' and 'sku' to include the main SKU ids which each SKU belongs too
aznUS.insert(3,'group-sku', '')
```

```python
aznUS
```

```python
# Inserting the group-skus 
aznUS['sku'] = aznUS['sku'].str.upper() 
for i in range(0,len(aznUS['sku'].unique())):
    for j in range(0,len(databaseLandingCosts['Sub-ID'].unique())):
            if aznUS['sku'].unique()[i] == databaseLandingCosts['Sub-ID'][j]:
                aznUS.loc[aznUS['sku']== aznUS['sku'].unique()[i],'group-sku'] = databaseLandingCosts['ID Group'][j]
```

```python
# Adding in the landing costs
aznUS['landing cost'] = '0'

for i in range(0,len(aznUS['sku'].unique())):
    for j in range(0,len(databaseLandingCosts['Sub-ID'].unique())):
        if aznUS['sku'].unique()[i] == databaseLandingCosts['Sub-ID'][j]:
            aznUS.loc[aznUS['sku']==aznUS['sku'].unique()[i],'landing cost'] = databaseLandingCosts['landing cost'][j]
```

```python
aznUS
```

```python
aznUS.dtypes
aznUS['landing cost'] = pd.to_numeric(aznUS['landing cost'])
```

```python
# In this cell i am adjusting the landing costs with respect to the amount of quantity bought from that record
for i in range(0,len(aznUS['sku'])):
    aznUS['landing cost'][i] = aznUS['landing cost'][i]*aznUS['quantity'][i]
    #print(aznUS['landing cost'][i])
aznUS
```

```python
# Categorizing Data 
aznUS = pd.read_csv("usReportPreAgg.csv")
```

```python
aznUS
```

```python
aznUS['selling fees'] = -(aznUS['selling fees'])
aznUS['fba fees'] = -(aznUS['fba fees'])
```

```python
# Extracting all the type Order records. This is because these are the actual transactions which have gone through
aznUSOrders = aznUS.loc[aznUS['type']=="Order"]
```

```python
aznUSOrders['sku'].unique()
aznUSAgg = []
for i in aznUSOrders['sku'].unique():
     aznUSAgg.append(i)
```

```python
aznUSAgg = pd.DataFrame(aznUSAgg)
```

```python
aznUSAgg = aznUSAgg.rename(columns = {0:"sku"})
```

```python
aznUSAgg['Total-Quantity-Sold'] = ''
aznUSAgg['Total-Product-Revenue'] = ''
aznUSAgg['Total-Selling-Fee'] = ''
aznUSAgg['Total-Fba-Fees'] = ''
aznUSAgg['Total-Landing-Costs'] = ''
aznUSAgg['Total-Product-Profit'] = ''
aznUSAgg['R.O.I'] = ''
aznUSAgg['Profit Per Quantity Sold'] = ''
```

```python
aznUSOrders = aznUSOrders.groupby('sku').agg({'quantity':'sum','product sales':'sum','selling fees':'sum','fba fees':'sum','landing cost':'sum'}).reset_index()
```

```python
# Sorting and reseting indexes and dropping

# aznUSAgg = aznUSAgg.sort_values(by ='sku', ascending = True).reset_index()
# aznUSAgg = aznUSAgg.drop(columns = ['index'])
```

```python
for i in range(0,len(aznUSAgg['sku'])):
    for j in range(0,len(aznUSOrders['sku'])):
        if  aznUSAgg['sku'][i] == aznUSOrders['sku'][j]:
            aznUSAgg['Total-Quantity-Sold'][i] = aznUSOrders['quantity'][j]
            aznUSAgg['Total-Product-Revenue'][i] = aznUSOrders['product sales'][j]
            aznUSAgg['Total-Selling-Fee'][i] = aznUSOrders['selling fees'][j]
            aznUSAgg['Total-Fba-Fees'][i] = aznUSOrders['fba fees'][j]
            aznUSAgg['Total-Landing-Costs'][i] = aznUSOrders['landing cost'][j]
            aznUSAgg['Total-Product-Profit'][i] = ''
            aznUSAgg['R.O.I'][i] = ''
            aznUSAgg['Profit Per Quantity Sold'][i] = ''
```

```python
# Calculating the profit for that item 
aznUSAgg['Total-Product-Profit'] = aznUSAgg['Total-Product-Revenue'] - (aznUSAgg['Total-Selling-Fee'] +aznUSAgg['Total-Fba-Fees']+aznUSAgg['Total-Landing-Costs'])
# Calculating the profit per quantity sold 
aznUSAgg['Profit Per Quantity Sold'] = aznUSAgg['Total-Product-Profit']/aznUSAgg['Total-Quantity-Sold']
# Calculating the R.O.I (Actually Sales)
aznUSAgg['R.O.I'] = aznUSAgg['Total-Product-Profit']/aznUSAgg['Total-Quantity-Sold']
```

```python
aznUSAgg = aznUSAgg.rename(columns = {"R.O.I":"Return on Sales"})
```

```python
import numpy as np
import pandas as pd
from datetime import datetime 
import matplotlib.pylab as plt
%matplotlib inline 
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,10
```

```python
data = pd.read_csv(r"net32Top1ReadyForTSA.csv")
data = data.iloc[7:]
```

```python
data
```

```python
data['date']= pd.to_datetime(data['date'], infer_datetime_format = True)
```

```python
data = data.set_index("date")
```

```python
data = data[['sumOfProfitNet32']]
data
```

```python
plt.xlabel('Date')
plt.ylabel('7DayAvgReturnOnSales')
plt.title('7DayAvgReturnOnSales over time')
plt.plot(data)
plt.xticks(rotation = 45)
plt.subplots_adjust(bottom=0.15)
```

```python
# Rolling Statistics Mean
rollingMeanSumOfProfit = data.rolling(window=12).mean()

print(rollingMeanSumOfProfit)
```

```python
# Rolling Statistics Standard Deviation
rollingSTDSumOfProfit = data.rolling(window =12).std()
print(rollingSTDSumOfProfit)
```

```python
# Plotting the rolling statistics 
rcParams['figure.figsize'] = 10,10

dataGraph = plt.plot(data, color = 'blue', label = 'Original')
rolligMeanGraph = plt.plot(rollingMeanSumOfProfit, color = 'red', label = 'Rolling Mean')
rollingSTDGraph = plt.plot(rollingSTDSumOfProfit, color = 'black', label = 'Rolling STD')
```

```python
# Preform the dickey-fuller test: 
from statsmodels.tsa.stattools import adfuller 

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(data['sumOfProfitNet32'],autolag = 'AIC')

dfoutput = pd.Series(dftest[0:4], index = ['Test Statistic', 'p-value', '#Lags Used','Number of Observations Used'])

for key,value in dftest[4].items():
    dfoutput['Crticial Value (%s)'%key] = value 

print(dfoutput)
```

```python
#Estimating trend  with log 
ProfitDatalogScale = np.log(data)
plt.plot(ProfitDatalogScale)
```

```python
datalogScaleMovingAverage = ProfitDatalogScale.rolling(window=12).mean()
datalogScaleSTD = ProfitDatalogScale.rolling(window =12).std()

plt.plot(ProfitDatalogScale)
plt.plot(datalogScaleMovingAverage, color = 'red')
plt.plot(datalogScaleSTD, color = 'blue')
```

```python
logDataMinuslogMovingAverage = ProfitDatalogScale - datalogScaleMovingAverage
logDataMinuslogMovingAverage
```

```python
#removing NAN values 
logDataMinuslogMovingAverage.dropna(inplace = True)
logDataMinuslogMovingAverage
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity (timeseries):
    
    #Determing rolling statistics 
    movingAverage = timeseries.rolling(window =12).mean()
    movingSTD = timeseries.rolling(window =12).std()
    
    #Plot rolling statistics 
    orig = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(movingAverage, color = 'red', label = 'Rolling Mean')
    std = plt.plot(movingSTD, color = 'black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Meand & STD')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries['sumOfProfitNet32'], autolag= 'AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lags used','Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Crticial Value (%s)'%key] = value 
        
    print(dfoutput)
```

```python
test_stationarity(logDataMinuslogMovingAverage)
```

```python
#The weighted Average of Time-Series
exponentialDecayWeightedAverage = ProfitDatalogScale.ewm(halflife = 12, min_periods = 0 , adjust = True).mean()
plt.plot(ProfitDatalogScale)
plt.plot(exponentialDecayWeightedAverage, color = 'red')
```

```python
profitLogMinusMovingExponentialDecayAverage =  ProfitDatalogScale - exponentialDecayWeightedAverage
profitLogMinusMovingExponentialDecayAverage = profitLogMinusMovingExponentialDecayAverage.fillna(0)
test_stationarity(profitLogMinusMovingExponentialDecayAverage)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ProfitDatalogScale)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ProfitDatalogScale, label = 'Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label = 'Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label = 'Residual')
plt.legend(loc='best')
plt.tight_layout()

decomposedLogData = residual 
decomposedLogData.dropna(inplace=True)
test_stationarity(decomposedLogData)
```

```python
decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
test_stationarity(decomposedLogData)
```

```python
# ACF graph and PACF plots: 
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(datasetLogDiffShifting, nlags=20)
lag_pacf = pacf(datasetLogDiffShifting, nlags = 20, method= 'ols')

# Plot ACF: AutoCorrelation function: 
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle = '--', color = 'gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShifting)), linestyle = '--', color = 'gray')
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--', color = 'gray')
plt.title('Autocorrelation Function')

#Plot PACF: Partial Autocorrelation Function: 

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle = '--', color = 'gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShifting)), linestyle = '--', color = 'gray')
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--', color = 'gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR MODEL
model = ARIMA(ProfitDatalogScale, order = (2,1,0))
results_AR = model.fit(disp=-1)
plt.plot(datasetLogDiffShifting)
plt.plot(results_AR.fittedvalues, color = 'red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-datasetLogDiffShifting["sumOfProfitNet32"]).fillna(0)**2))
plt.legend(loc='best')
print('plotting AR model')
```

```python
# MA model
model = ARIMA(ProfitDatalogScale, order =(0,1,2))
results_MA = model.fit(disp=-1)
plt.plot(datasetLogDiffShifting)
plt.plot(results_MA.fittedvalues, color = 'red')
plt.title('RSS:%.4f'% sum((results_MA.fittedvalues-datasetLogDiffShifting['sumOfProfitNet32']).fillna(0)**2))
print('Plotting AR model')
```

```python
# ARIMA model:

model = ARIMA(ProfitDatalogScale, order = (2,1,2))
results_ARIMA = model.fit(disp=-1)
plt.plot(datasetLogDiffShifting)
plt.plot(results_ARIMA.fittedvalues, color = 'red')
plt.title('RSS:%.4f'% sum((results_ARIMA.fittedvalues-datasetLogDiffShifting['sumOfProfitNet32']).fillna(0)**2))
```

```python
# Creating the prediction 

predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy = True)
print(predictions_ARIMA_diff)
```

```python
# Convert to cumulative sum 
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum)
```

```python
predictions_ARIMA_log = pd.Series(ProfitDatalogScale['sumOfProfitNet32'].ix[0], index=ProfitDatalogScale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value = 0)
predictions_ARIMA_log
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(data)
plt.plot(predictions_ARIMA)
```

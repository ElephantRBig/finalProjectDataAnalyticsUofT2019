```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dates = pd.date_range(start = "1/1/2019", end = "11/04/2019")
dates
```

```python
data = pd.read_csv(r'top1BestSellingAmazonPredictionOMR-BP742N.csv')
data
```

```python
data['date'] = data['date'].astype(str)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
dates = pd.DataFrame(dates)
dates = dates.rename(columns = {0:'date'})
```

```python
# dates = pd.DataFrame(dates)
# dates = dates.rename(columns = {0:'date'})
# data.columns
list = ['Profit_OMR-BP742N', 'Avg. return on sales_OMR-BP742N',
       'Distinct count of order id_OMR-BP742N', 'quantity_OMR-BP742N',
       'profitPerQuantitySold_OMR-BP742N', 'profitPerOrderMade_OMR-BP742N',
       'Total Cost_OMR-BP742N', 'Prediction_OMR-BP742N']
```

```python
for i in list:
    dates[f'{i}']= ' '
```

```python
for i in range(0,len(dates['date'])):
    for j in range(0,len(data['date'])):
        if dates['date'][i] == data['date'][j]:
            dates['Profit_OMR-BP742N'][i] = data['Profit_OMR-BP742N'][j]
            dates['Avg. return on sales_OMR-BP742N'][i]= data['Avg. return on sales_OMR-BP742N'][j]
            dates['Distinct count of order id_OMR-BP742N'][i]= data['Distinct count of order id_OMR-BP742N'][j]
            dates['quantity_OMR-BP742N'][i]= data['quantity_OMR-BP742N'][j]
            dates['profitPerQuantitySold_OMR-BP742N'][i]= data['profitPerQuantitySold_OMR-BP742N'][j]
            dates['profitPerOrderMade_OMR-BP742N'][i]= data['profitPerOrderMade_OMR-BP742N'][j]
            dates['Total Cost_OMR-BP742N'][i] = data['Total Cost_OMR-BP742N'][j]
            dates['Prediction_OMR-BP742N'][i] = data['Prediction_OMR-BP742N'][j]
            continue
```

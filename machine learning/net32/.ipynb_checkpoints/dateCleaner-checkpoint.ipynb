```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dates = pd.date_range(start = "1/1/2019", end = "11/04/2019")
dates
```

```python
data = pd.read_csv(r'top1BestSellingItems_Net32.csv')
data
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
dates = pd.DataFrame(dates)
dates = dates.rename(columns = {0:'date'})
```

```python
# dates = pd.DataFrame(dates)
# dates = dates.rename(columns = {0:'date'})
# data.columns
list = ['Profit_3MD-712-036', 'Avg. return on sales_3MD-712-036',
       'Distinct count of order id_3MD-712-036', 'quantity_3MD-712-036',
       'profitPerQuantitySold_3MD-712-036', 'profitPerOrderMade_3MD-712-036', 'Prediction_3MD-712-036']
```

```python
list
for i in list:
    dates[f"{i}"] = ' '
```

```python
data
```

```python
for i in range(0,len(dates['date'])):
    for j in range(0,len(data['date'])):
        if dates['date'][i] == data['date'][j]:
            dates['Profit_3MD-712-036'][i] = data['Profit_3MD-712-036'][j]
            dates['Avg. return on sales_3MD-712-036'][i]= data['Avg. return on sales_3MD-712-036'][j]
            dates['Distinct count of order id_3MD-712-036'][i]= data['Distinct count of order id_3MD-712-036'][j]
            dates['quantity_3MD-712-036'][i]= data['quantity_3MD-712-036'][j]
            dates['profitPerQuantitySold_3MD-712-036'][i]= data['profitPerQuantitySold_3MD-712-036'][j]
            dates['profitPerOrderMade_3MD-712-036'][i]= data['profitPerOrderMade_3MD-712-036'][j]
            dates['Prediction_3MD-712-036'][i] = data['Prediction_3MD-712-036'][j]
            continue
```

```python
dates.to_csv(r'net32Top1ReadyForML.csv')
```

```python
# Dropping an unnecessary column 
net32Report = net32Report.drop(columns = {'Sales Price'})

# Renaming a column to properly represent it's content
net32Report = net32Report.rename(columns = {'Amount':'Sales Total'})

# Dropping an unnecessary column 
net32Report = net32Report.drop(columns = {'Ship To Address 1'})

# Renaming a column to properly represent it's content
net32Report = net32Report.rename(columns = {'Ship To Address 2':'Address'})

# Renaming a column to properly represent it's content
net32Report = net32Report.rename(columns ={'Num':'Invoice Number'})
```

```python
# In this cell i will be importing the cost of goods for net32 items 
cost = pd.read_csv('net32CostOfGoods.csv')
```

```python
cost.head(1)
```

```python
# Adding in the landing costs
net32Report['landing cost'] = '0'

for i in range(0,len(net32Report['Item'].unique())):
    for j in range(0,len(cost['Item'].unique())):
        if net32Report['Item'].unique()[i] == cost['Item'][j]:
            net32Report.loc[net32Report['Item']==net32Report['Item'].unique()[i],'landing cost'] = cost['Cost'][j]
```

```python
net32Report.dtypes
net32Report['landing cost'] = pd.to_numeric(net32Report['landing cost'])
```

```python
# In this cell i am adjusting the landing costs with respect to the amount of quantity bought from that record
for i in range(0,len(net32Report['Item'])):
    net32Report['landing cost'][i] = net32Report['landing cost'][i]*net32Report['Qty'][i]
    #print(net32Report['landing cost'][i])
net32Report
```
